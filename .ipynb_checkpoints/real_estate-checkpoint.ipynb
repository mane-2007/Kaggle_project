```python
cols_to_keep = [
    'zip', 'year_built', 'sqft', 'stories', 
    'beds', 'baths', 'garage', 'lastSoldPrice','type'
]
df_clone = df[cols_to_keep].copy()
df_clone = df_clone.dropna(subset=['lastSoldPrice'])  #I dropped the NA values of lastSoldPrice because it was only 15 rows which is not a lot compared to 12k rows
print(f"Original shape: {df.shape}")
print(f"Clone shape: {df_clone.shape}")
print(df_clone.head())
```

```python
#Specifically fill 'garage' with 0
df_clone['garage'] = df_clone['garage'].fillna(0)

#Fill any remaining NaNs with the Median
#This ensures columns like beds/baths/sqft get typical values, not 0s.
df_clone = df_clone.fillna(df_clone.median(numeric_only=True))

#ONE-HOT ENCODE the 'type' column
#This turns 'type' into columns like 'type_condo', 'type_house', etc.
df_clone = pd.get_dummies(df_clone, columns=['type'], prefix='type')

#Verify there are no more missing values
print(df_clone.isnull().sum())
print(df_clone.head())
print(df_clone.shape)
print(df.shape)
```

```python
#Create the region (Step 1)
df_clone['region'] = df_clone['zip'].astype(str).str[:3]

#Convert ZIP to a string so the model treats it as a category, not a number (Step 2)
df_clone['zip_cat'] = df_clone['zip'].astype(str)

#One-Hot Encode BOTH
# This captures the broad region AND the specific neighborhood
df_clone = pd.get_dummies(df_clone, columns=['region', 'zip_cat'], prefix=['reg', 'zip'])

#drop the original numerical zip
df_clone = df_clone.drop(columns=['zip'])

print(f"New Shape: {df_clone.shape}") 
```

```python
df_clone.tail()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#Prepare the Price Classes (as required by rubric)
df_clone['price_class'] = pd.qcut(df_clone['lastSoldPrice'], q=4, labels=['Budget', 'Standard', 'High-End', 'Luxury'])

#Create a 2x2 grid of plots
plt.figure(figsize=(16, 12))

#Plot 1: Target Price Distribution ---
plt.subplot(2, 2, 1)
plt.hist(df_clone['lastSoldPrice'], bins=50, color='teal', edgecolor='white')
plt.yscale('log') # Handles those $15M outliers
plt.title('Log Distribution of Last Sold Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency (Log Scale)')

#Plot 2: SQFT Histograms by Price Class (Rubric Requirement) ---
plt.subplot(2, 2, 2)
colors = ['gray', 'blue', 'orange', 'red']
for i, p_class in enumerate(['Budget', 'Standard', 'High-End', 'Luxury']):
    subset = df_clone[df_clone['price_class'] == p_class]
    plt.hist(subset['sqft'], bins=30, alpha=0.5, label=p_class, color=colors[i], density=True)
plt.title('SQFT Distribution per Price Tier')
plt.xlabel('Square Footage')
plt.ylabel('Density')
plt.legend()

#Plot 3: Scatter Plot (Price vs SQFT) ---
plt.subplot(2, 2, 3)
plt.scatter(df_clone['sqft'], df_clone['lastSoldPrice'], alpha=0.2, s=5, color='purple')
plt.yscale('log')
plt.title('Price vs. Square Footage (Trend Analysis)')
plt.xlabel('SQFT')
plt.ylabel('Price ($)')

#Plot 4: Price by Bedroom Count ---
plt.subplot(2, 2, 4)
# Filter for 0-6 beds to keep the chart readable
bed_counts = sorted([b for b in df_clone['beds'].unique() if b <= 6])
bed_data = [df_clone[df_clone['beds'] == b]['lastSoldPrice'] for b in bed_counts]
plt.boxplot(bed_data, labels=[int(b) for b in bed_counts])
plt.yscale('log')
plt.title('Price Variance by Bedroom Count')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price ($)')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score

# 1. Define X and y (Dropping the 'cheating' columns)
X = df_clone.drop(columns=['lastSoldPrice', 'price_class'])
y = df_clone['lastSoldPrice']

# 2. Split into Train (70%), Val (15%), and Test (15%)
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.176, random_state=42)

# 3. Rescale everything
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# 4. Train the "Two-Step" Model (Ridge Regression)
model = Ridge(alpha=1.0) # alpha helps prevent the 505 features from overfitting
model.fit(X_train_scaled, y_train)

# 5. Check the initial performance on the Validation set
y_pred = model.predict(X_val_scaled)
mae = mean_absolute_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)

print(f"Mean Absolute Error: ${mae:,.2f}")
print(f"R-squared Score: {r2:.4f}")
```

```python
import numpy as np

# 1. Transform the target to Log scale
# np.log1p is safer than np.log because it handles 0s correctly
y_log = np.log1p(df_clone['lastSoldPrice'])

# 2. Re-split with the log target
X = df_clone.drop(columns=['lastSoldPrice', 'price_class'])
X_train_v, X_test, y_train_v, y_test = train_test_split(X, y_log, test_size=0.15, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train_v, y_train_v, test_size=0.176, random_state=42)

# 3. Scale and Train
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

model_log = Ridge(alpha=1.0)
model_log.fit(X_train_scaled, y_train)

# 4. Predict and transform BACK to dollars to check error
y_pred_log = model_log.predict(X_val_scaled)
y_pred_dollars = np.expm1(y_pred_log)
y_val_dollars = np.expm1(y_val)

# 5. Calculate Metrics
mae_dollars = mean_absolute_error(y_val_dollars, y_pred_dollars)
r2_log = r2_score(y_val, y_pred_log) # R2 on log scale is usually much higher

print(f"New Mean Absolute Error: ${mae_dollars:,.2f}")
print(f"New R-squared Score (Log Scale): {r2_log:.4f}")

# Calculate Median Percentage Error (Very useful for your presentation!)
mape = np.median(np.abs((y_val_dollars - y_pred_dollars) / y_val_dollars)) * 100
print(f"Median Percentage Error: {mape:.2f}%")
```
